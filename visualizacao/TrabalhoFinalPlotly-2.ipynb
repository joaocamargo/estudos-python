```python
df = pd.read_csv('https://raw.githubusercontent.com/joaocamargo/estudos-python/master/visualizacao/dados_municipais.csv')
```

```python
df['PERCENT_ESCOLA_PUB'] = ((df['MAT_TOTAL']*100)/  (df['MAT_TOTAL'] + df['MAT_PART']))
df['PERCENT_ESCOLA_PART'] = ((df['MAT_PART']*100)/  (df['MAT_TOTAL'] + df['MAT_PART']))
```

# Normalizar colunas

```python
#criar colunas a normalizar
#normalized_df=(df-df.mean())/df.std()

df['EDUCACAO_NORM'] = (df['EDUCACAO']-df['EDUCACAO'].min())/(df['EDUCACAO'].max()-df['EDUCACAO'].min())
df['SAUDE_NORM'] = (df['SAUDE']-df['SAUDE'].min())/(df['SAUDE'].max()-df['SAUDE'].min())
df['MAT_INF_NORM'] = (df['MAT_INF']-df['MAT_INF'].min())/(df['MAT_INF'].max()-df['MAT_INF'].min())
df['MAT_FUN_NORM'] = (df['MAT_FUN']-df['MAT_FUN'].min())/(df['MAT_FUN'].max()-df['MAT_FUN'].min())
df['MAT_TOTAL_NORM'] = (df['MAT_TOTAL']-df['MAT_TOTAL'].min())/(df['MAT_TOTAL'].max()-df['MAT_TOTAL'].min())
df['MAT_PART_NORM'] = (df['MAT_PART']-df['MAT_PART'].min())/(df['MAT_PART'].max()-df['MAT_PART'].min())
df['DESP_POR_ALUNO_NORM'] = (df['DESP_POR_ALUNO']-df['DESP_POR_ALUNO'].min())/(df['DESP_POR_ALUNO'].max()-df['DESP_POR_ALUNO'].min())
display(df.head(5))
```
